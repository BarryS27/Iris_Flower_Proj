```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")
```

```python
df.head()
```

```python
df.value_counts(df.species)
```

```python
avg_sepal_length = df.groupby("species")["sepal_length"].mean().reset_index()
avg_sepal_width = df.groupby("species")["sepal_width"].mean().reset_index()
avg_petal_length = df.groupby("species")["petal_length"].mean().reset_index()
avg_petal_width = df.groupby("species")["petal_width"].mean().reset_index()
```

```python
species = {
"setosa": df[df["species"] == "setosa"],
"versicolor": df[df["species"] == "versicolor"],
"virginica": df[df["species"] == "virginica"]
}
colors = ["pink","tan","green"]
```

```python
for i in species,colors:
  plt.scatter(species[i].sepal_length,species[i].sepal_width,c=colors[i],s=16)
```

```python
plt.subplot(2,2,1)
plt.scatter(setosa.sepal_length,setosa.sepal_width,c="tan",s=16)
plt.scatter(versicolor.sepal_length,versicolor.sepal_width,c="pink",s=16)
plt.scatter(virginica.sepal_length,virginica.sepal_width,c="green",s=16)
```
